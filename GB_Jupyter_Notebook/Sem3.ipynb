```python
# 256GB SSD +  1.0TB Hybrid
def get_gb(row):
    row_list = row.split()
    gb=0
    for elem in row_list:
            if "GB" in elem: 
                gb+=float(elem[:-2])
            elif 'TB' in elem:
                gb+=float(elem[:-2])*1024
    return int(gb)
    
get_gb('256GB SSD +  1.0TB Hybrid') #тест строки 1024 + 256


    
    
```

```python
df['Memory_Amount'] = df['Memory'].apply(get_gb)
df.head()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)


```python
df['Memory'].value_counts()
```

```python
# 64GB Flash Storage +  1TB HDD 
def type_memory1(row):
    row_list = row.split()
    type=''
    for elem in row_list:
        if "GB" not in elem and "TB" not in elem: 
                type+=elem + ' '
    return type.rstrip() # удаляет пробелы в конце и табуляции
type_memory1("64GB Flash Storage +  1TB HDD") 
```

```python
# 64GB Flash Storage +  1TB HDD 
def type_memory2(row):
    row_list = row.split()
    type=''
    for elem in row_list:
        if elem.isalpha(): # строка из букв, если есть цифры или +- - убирает
                type+=elem + ' '
    return type.rstrip() # удаляет пробелы в конце и табуляции
type_memory2("64GB Flash Storage +  1TB HDD")
```

```python
df['Memory_Type1'] = df['Memory'].apply(type_memory1) 
df['Memory_Type2'] = df['Memory'].apply(type_memory2)
df.head()
```

```python
df['Memory_Type1'].value_counts()
```

```python
df['Memory_Type2'].value_counts()
```
